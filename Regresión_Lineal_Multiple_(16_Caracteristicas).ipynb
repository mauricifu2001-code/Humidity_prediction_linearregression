# Homocedasticidad: Igualdad de varianzas

```python
sns.regplot(x=fitted_val, y=resid_val, lowess=True, line_kws={'color':'red'})  # Valores ajustados Vs Residuales (residuales puros)

plt.title('Fitted vrs Residual')

#
```

```python
# Es mejor utilizar valores residuales estandarizados:

resid_stand = lin_reg.get_influence().resid_studentized_internal  # Estandarizar los valores residuales

sns.regplot(x=fitted_val, y=resid_stand, lowess=True, line_kws={'color':'red'})
plt.title('Fitted vrs Residual')
```

```python
import statsmodels as sms
# Test de Breush-Pagan para saber si hay homocedasticidad:

bp_test = sms.stats.diagnostic.het_breuschpagan(resid_val, lin_reg.model.exog)

print(bp_test)

print("\nBreush-Pagan test: pvalue = ",bp_test[1])
```

El pvalue es menor a 0.05 por lo que rechazo la hipotesis nula y cual es la hipotesis nula? es igualdad de varianza y la hipotesis alternativa es que la distribucion de mis residuales no tienen una varianza constante.

por lo que el modelo no cumple ni la linealidad y ni la homocedasticidad. Por lo que toca corregir todo eso.


**Heterocedasticidad = No hay una varianza constante == Malo**

# Outliers elimination for dependent variable:

```python
# Finding the Boundary Values

p3 = y.mean() + 3*y.std()
p1 = y.mean() - 3*y.std()

print("Highest allowed",p3)
print("Lowest allowed",p1)

#Finding the Outliers
print( X[(y > p3) | (y < p1)])
```

```python
# Finding the IQR

percentile25 = y.quantile(0.25) #Q1
percentile75 = y.quantile(0.75) #Q2
iqr = percentile75 - percentile25

# Finding upper and lower limit

upper_limit = percentile75 + 1.5*iqr
lower_limit = percentile25 - 1.5*iqr
```
